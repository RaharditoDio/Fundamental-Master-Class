```python
import numpy as np
x = [10,20,30,40,50]
y = np.array(x)**2
```

```python
plt.plot(x,y);
```

```python
plt.plot(x,y)
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Judul Grafik');
```

```python
plt.subplot(121)
plt.plot(x,y,color = 'red')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.subplot(122)
plt.plot(x,y,color = 'blue')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.show()
```

```python
plt.subplot(121)
plt.plot(x,y,color = 'red')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.subplot(122)
plt.plot(x,y,color = 'blue')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.tight_layout()
plt.show()
```

```python
x = np.arange(1,20,0.5)
y = x**2
```

```python
# versi 1 (matplotlib OOP)
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x,y)
axes.set_xlabel('Sumbu X')
axes.set_ylabel('Sumbu Y')
axes.set_title('Judul Grafik')
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])

axes1.plot(x,y)
axes1.set_xlabel('Sumbu X besar')
axes1.set_ylabel('Sumbu Y besar')
axes1.set_title('Judul Grafik Utama')

axes2.plot(y,x)
axes2.set_xlabel('Sumbu X kecil')
axes2.set_ylabel('Sumbu Y kecil')

plt.show()
```

```python
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[0].set_xlabel('Sumbu X1')
axes[0].set_ylabel('Sumbu Y1')
axes[0].set_title('Judul 1')

axes[1].plot(y,x)
axes[1].set_xlabel('Sumbu X2')
axes[1].set_ylabel('Sumbu Y2')
axes[1].set_title('Judul 2')

fig.tight_layout()
```

```python
a = np.arange(1,20,0.5)
b = a**2
```

```python
fig = plt.figure(figsize = (4,4), dpi = 100)
ab = fig.add_axes([0,0,1,1])
ab.plot(a,b);
```

```python
fig, axes = plt.subplots(2,1, figsize = (4,4), dpi = 70)
axes[0].plot(a,b)
axes[0].set_title('Grafik 1')

axes[1].plot(a,b)
axes[1].set_title('Grafik 2')
plt.tight_layout()

plt.show()
```

## Menambah Legenda

```python
fig = plt.figure(facecolor = 'yellow', dpi = 80)
ax = fig.add_axes([0,0,1,1])
ax.plot(a,b, label = 'Pangkat 2')
ax.plot(a,a**3, label = 'Pangkat 3')
ax.legend(loc=(0.5, 0.5))
plt.show()
```

```python
fig = plt.figure(facecolor = 'yellow', dpi = 80)
ax = fig.add_axes([0,0,1,1])
ax.plot(a,b, label = 'Pangkat 2')
ax.plot(a,a**3, label = 'Pangkat 3')
ax.plot(a,a**4, label = 'Pangkat 4')
ax.legend(loc=(0.5, 0.5))
ax.grid(True)
plt.show()
```

```python
fig = plt.figure(facecolor = 'yellow', dpi = 80)
ax = fig.add_axes([0,0,1,1])

ax.plot(a,b, label = 'Pangkat 2', ls = '--')
ax.plot(a,a**3, label = 'Pangkat 3', lw = '5')
ax.plot(a,a**4, label = 'Pangkat 4')
ax.legend(loc=(0.5, 0.5))

ax.set_xlim([0,12.5])
ax.set_ylim([0,2500])

ax.grid(True)
plt.show()
```
