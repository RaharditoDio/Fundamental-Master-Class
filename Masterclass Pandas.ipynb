## Merge

```python
one = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                        'one':[1,2,3,4,5],
                        'two':[6,7,8,9,10]})
```

```python
two = pd.DataFrame({'key':['k0','k1','k2','k3','k4'],
                        'three':[1,2,3,4,5],
                        'four':[6,7,8,9,10]})
```

```python
one
```

```python
two
```

```python
onetwo = pd.merge(one, two, how = 'inner', on = 'key')
onetwo
```

```python
key_A = pd.DataFrame({'key1':['K0','K0','K1','K2'],
                      'key2':['K0','K1','K0','K1'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3']})
                     
key_B = pd.DataFrame({'key1':['K0','K1','K1','K2'],
                      'key2':['K0','K0','K0','K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
key_A
```

```python
key_B
```

```python
keyleft = pd.merge(key_A, key_B, how = 'left', on = ['key1','key2'])
keyleft
```

```python
keyright = pd.merge(key_A, key_B, how = 'right', on = ['key1','key2'])
keyright
```

```python
keyouter = pd.merge(key_A, key_B, how = 'right', on = ['key1','key2'])
keyouter
```

## JOIN

```python
testA = pd.DataFrame({'A':['A0','A1','A2'],
                     'B':['B0','B1','B2']},
                    index = ['K0','K1','K2'])

testB = pd.DataFrame({'C':['C0','C2','C3'],
                     'D':['D0','D2','D3']},
                    index = ['K0','K1','K2'])
```

```python
testA
```

```python
testB
```

```python
testA.join(testB)
```

```python
testA.join(testB, how='inner')
```
