## Visualization Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
dataset = pd.read_csv('kapal_titanic.csv')
```

```python
dataset.head()
```

```python
dataset.describe(include = 'O')
```

```python
# barplot (histogram)
sns.barplot(x = 'sex', y = 'fare', data = dataset);
```

```python
# barplot (histogram)
sns.barplot(x = 'fare', y = 'sex', data = dataset);
```

```python
# barplot (histogram)
sns.barplot(x = 'sex', y = 'fare', data = dataset, estimator = sum);
```

```python
# barplot (histogram)
sns.barplot(x = 'sex', y = 'fare', data = dataset, estimator = np.std);
```

```python
# countplot (histogram)
sns.countplot(x = 'sex', data = dataset);
```

```python
sns.countplot(x = 'deck', hue = 'sex', data=dataset);
```

```python
sns.countplot(x = 'deck', hue = 'sex', order = ['A','B','C','D','E','F','G'], data=dataset);
```

```python
# boxplot (box & whiskers plot)
sns.boxplot(x = 'sex', y = 'fare', data = dataset);
```

```python
# boxplot (box & whiskers plot)
sns.boxplot(x = 'sex', y = 'fare', hue = 'deck', hue_order = ['A','B','C','D','E','F','G'], data = dataset);
```

```python
# violin plot
sns.violinplot(x = 'sex', y = 'fare', data = dataset);
```

```python
# violinplot (box & whiskers plot)
sns.violinplot(x = 'sex', y = 'fare', hue = 'deck', hue_order = ['A','B','C','D','E','F','G'], data = dataset);
```

```python
sns.violinplot(x = 'deck', y = 'fare', hue = 'sex', order = ['A','B','C','D','E','F','G'], split = True, data = dataset);
```

```python
# Stripplot
sns.stripplot(x = 'deck', y = 'fare', hue = 'sex', order = ['A','B','C','D','E','F','G'], jitter = True, data = dataset);
```

```python
# Swarmplot
sns.swarmplot(x = 'deck', y = 'fare', hue = 'sex', order = ['A','B','C','D','E','F','G'], data = dataset);
```

```python
# Menggabungkan Voilin & Swarmplot
sns.violinplot(x = 'deck', y = 'fare', hue = 'sex', order = ['A','B','C','D','E','F','G'], data = dataset);
sns.swarmplot(x = 'deck', y = 'fare', hue = 'sex', order = ['A','B','C','D','E','F','G'], color = 'black', data = dataset);
```

## FacetGrid

```python
tips = sns.load_dataset('tips')
tips
```

```python
facettips = sns.FacetGrid(tips, row = 'smoker', col = 'day')
facettips.map(sns.distplot, 'total_bill')
plt.show()
```

```python
facettips = sns.FacetGrid(tips, row = 'smoker', col = 'time')
facettips.map(sns.distplot, 'total_bill')
plt.show()
```

```python
facettips = sns.FacetGrid(tips, row = 'smoker', col = 'time')
facettips.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```

```python
facettips = sns.FacetGrid(tips, row = 'smoker', col = 'time', hue = 'sex')
facettips.map(plt.scatter, 'total_bill', 'tip')
plt.show()
```
